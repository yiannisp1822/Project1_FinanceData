```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load the CPIAUCSL data
cpi_df = pd.read_csv('./resources/CPIAUCSL.csv')
print("CPIAUCSL Data:")
display(cpi_df.head())

# Load the CIVPART data
civ_df = pd.read_csv('./resources/CIVPART.csv')
print("\nCIVPART Data:")
display(civ_df.head())
```

```python
# Plot CPIAUCSL data
plt.figure(figsize=(10, 5))
plt.plot(cpi_df['DATE'], cpi_df['CPIAUCSL'], label='CPIAUCSL', color='blue')
plt.xlabel('Date')
plt.ylabel('CPIAUCSL')
plt.title('CPIAUCSL Over Time')
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# Plot CIVPART data
plt.figure(figsize=(10, 5))
plt.plot(civ_df['DATE'], civ_df['CIVPART'], label='CIVPART', color='green')
plt.xlabel('Date')
plt.ylabel('CIVPART')
plt.title('CIVPART Over Time')
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```
