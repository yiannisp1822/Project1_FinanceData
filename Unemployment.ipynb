```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Read CSV
BLS = pd.read_csv('Resources/bls-monthly-unemployment-2000-2004.csv',
index_col = "Label",
parse_dates = True)

BLS.head()
```

```python
# Create Line Graph

# Plot the first few columns over time
BLS.plot(y=['Value'])
plt.title("Unemployment Rate")
plt.xlabel("Date")
plt.ylabel("Unemployment Rate")
plt.show()
```
